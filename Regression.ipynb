```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
path_test_regression = r'G:\Study\Statistics\test regression.csv'
data = pd.read_csv(path_test_regression)
```

```python
data
```

```python
data.describe()
```

```python
y = data['b']
x1 = data['a']
```

```python
plt.scatter(x1,y)
plt.xlabel('a',fontsize=20)
plt.ylabel('b',fontsize=20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat = 10.6151*x1 - 3542.9344
fig = plt.plot(x1,yhat,lw=4,c='orange',label='regression line')
plt.xlabel('a',fontsize=20)
plt.ylabel('b',fontsize=20)
plt.show()
```
