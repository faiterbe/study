```python
def prob(world, cur):
    return len(set.intersection(world,cur)) / len(world)

def prob_union(world,sets):
    counter = 0
    for cur in sets:
        counter += prob(world,cur)
    return counter

def prob_intersect(world,sets):
    sets.append(world)
    intersect = set()
    for cur in sets:
        if intersect:            
            intersect = set.intersection(intersect,cur)
        else:
            intersect = cur        
    return len(intersect) / len(world)

def prob_conditional(world,set_a,set_b):
    return prob_intersect(world,[set_a,set_b]) / prob(world,set_b)
```

```python
from fractions import Fraction

def bayes_predict(train,test):
    
    # (key,value) = ({0,1},Fraction)
    outcomes = dict()
    for value in train.values():
        outcomes[value] = Fraction(1,1)
    
    for outcome in outcomes:
        print('outcome[{0}]:'.format(outcome))
        for i in range(len(test)):
            print('i={0}'.format(i))
            count_all = 0
            count_value = 0
            for sample, outcome_train in train.items():
                print('{0} , {1}'.format(sample, outcome_train))
                print('{0} ?= {1}'.format(sample[i],test[i]))
                if outcome == outcome_train:                
                    count_all += 1
                    if sample[i] == test[i]:
                        count_value += 1
            outcomes[outcome] *= Fraction(count_value+1,count_all+2)
            print(outcomes[outcome])
        outcomes[outcome] *= Fraction(list(train.values()).count(outcome)+1,len(train.values())+2)
        print(outcomes[outcome])
            
```

```python
train = dict()
train[(0,0,0)] = 0
train[(0,0,1)] = 0
train[(1,1,0)] = 1
train[(1,1,1)] = 1

test = (0,1,1)

bayes_predict(train,test)
```
